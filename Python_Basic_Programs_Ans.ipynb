3. Using while loop accept numbers until sum of numbers is less than 100

```python
total_sum = 0
a = int(input())
while a != 0:
    total_sum += a
    if total_sum >= 100:
        print('Total sum is', total_sum)
        break
    a = int(input())
else:
    print('Total sum is less than 100 and is equal to', total_sum, '.')
```

5. Write a python program to scrape reviews from a commercial web site


```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = 'https://www.yelp.com/biz/milk-and-cream-cereal-bar-new-york?osq=Ice+Cream'
ourUrl = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(ourUrl,'html.parser')
```

```python
print(soup.prettify())
```

```python
review = []
for i in soup.find_all('div',{'class':'review-content'}):
  per_review=i.find('p')
  print(per_review)
  review.append(per_review)
```
